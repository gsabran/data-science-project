# <a id='overview'>3. Overview and Motivation</a>

The scientific and inferential goals for this project are to model and predict the temporal and geographic evolution of the Ebola outbreak. We would mainly like to better understand how the disease grows. If possible, we would also like to understand how it moves from one place to another and which covariates are most predictive of an outbreak (or associated with it). We implemented an SIR model in Python, which models a population across time as the number of susceptible individuals (S), number of infected individuals (I), and the number of removed individuals (R) changes. Our analysis is focused on the African Sierra Leone, Liberia, and Guinea. 

Our main finding is that the evolution of cases/deaths can be reasonably modeled using a SIR model, since it implicitly assumes an exponential growth of the number of cases/deaths. However, we identified that when trying to train the model to forecast evolution to cases at a regional level, the model does not have enough flexibility and significant modifications must be made to improve the fit.

This data science work was conducted in 3 main phases. We provide our high-level approach below and will describe the process in more detail in the following sections.

##B. Time series data

The sub-national time series data contained for each date and each regional unit, the number of cummulative cases and deaths. The caviat to this information is that it has large gaps of information for certain dates, multiple resources for each date, as well as inconsistencies in the number of cases between different geographical levels (for example, the aggregate of the numbers in counties do not add up to the number of the regions, which do not add up to the numbers on the national level). 

We took the following steps to ensure we had a workable timeseries dataset:

1. *Data format* We removed useless columns and rows. We also removed duplicate values and changed data types. Note: we decided to keep date format as an int (the way it was provided).

2. *Sources* We noticed the timeseries data had difference sources and observed that they were covering different time periods, but overlap stilled ocurred, so we prioritized the sources and filter our the data set to contain only the most reliable source for each date. We also standardized the source names.

3. *Time inconsistencies* We fixed the large gaps of information for certain dates by using linear interpolation. For example, Guinea has no sub-national data from June 20 to August 14 in 2014. We drew a straight line connecting the number of cases, deaths, and any other relevant numerical information to fill in the gaps between those dates so that we will have continuous data. We also removed inconsistencies in number of cases, deaths, and wherever we see large drops since by definition cases must be cummulative.

4. *Regional inconsistencies* We noticed inconsistencies in number of cases and deaths between different administrative levels. The time-series data that we have access to consisted of two levels of measurement at the ADM1 level (high-level regional) and at the ADM2 level (more fine-grained). We identified inconsistencies in the data where the ADM2 level did not correctly aggregate up to the ADM1 level but was consistent when aggregated to the national level. We spot-checked the lower-level ADM2 data. As a sanity check, we also double-checked with the official numbers released by news sources and verified that the numbers on the ADM1 level were not correct and the mistakes in those numbers was the source of the inconsistency. We disregarded the numbers on the ADM1 regional level and constructed the ADM1 level bottom-up, that is, from the numbers on the most granular level. Finally, by plotting the case count time-series for the data at the granular level we selected regions that had enough information to fit a model to (since we can't really fit models to regions that only have a few days of observations or very few Ebola cases).

5. *Recent Deaths* We constructed a variable containing the information for recent deaths for each date since while doing literature reviewed we noticed death body handling could be a potential factor affecting the spread of the disease.

```python
time_series = data[1].copy()
# manually identified wrong data 
bad_points_ids = [0, 1, 13639, 14131, 15653, 15995, 16337, 16716, 18394, 18490, 18586, 18682, 18778, 19342, 19438, 19534, 20500, 20596, 21940, 15655, 15997, 16339, 16732, 18410, 18506, 18602, 18698, 18794, 19358, 19454, 19550, 20516, 20612, 21956, 11203, 11389, 11575, 12169, 12379, 12649, 12925, 13213, 14005, 14341, 14677, 14889, 15083, 15377, 15469, 11511, 15475, 15481, 13641, 14133]
for i in bad_points_ids: time_series = time_series.drop(i)
# remove useless columns that prevent duplicates to be identified
time_series = time_series[['country_code', 'category', 'value', 'date', 'sources', 'sdr_id', 'sdr_name', 'sdr_level']]
time_series = time_series[(time_series.value == time_series.value) & (time_series.value != ' ')] # remove NaN values
time_series.value = time_series.value.astype(int) # convert values from string to int
time_series = time_series.drop_duplicates() # remove duplicates
time_series.head()
```

```python
#Standardize source name
def normalize_source(source):
    source = source.upper()
    if source == 'WHO; GVT': source = 'WHO'
    return source
time_series.sources = time_series.sources.apply(normalize_source)
time_series.sdr_level = time_series.sdr_level.fillna('national')
```

```python
# show the different sources for Guinea
def plot_raw_source(country_code, sdr_id):
    df = time_series[(time_series.country_code == country_code) & (time_series.sdr_id == sdr_id) & (time_series.category == 'Cases')]
    markers = ['x', 's', 'o', 'o', 'o', 'o', 'o', 'o', 'o', 'o']
    colors = ['b', 'r', 'g']
    for idx, s in enumerate(df.sources.unique()):
        _df = df[(df.sources == s)]
        plt.scatter(_df.date, _df.value, label=s, marker = markers[idx], color = colors[idx])
    display_date()
    plt.legend(loc=2)
def display_date():
    locs, labels = plt.xticks()
    plt.xticks(locs, map(lambda x: str((time_base + dt.timedelta(days=x)).date()), locs), rotation=45)

plt.title('Ebola cases for Guinea by source')
plot_raw_source('GN', 0)
```

```python
# transform the dataframe to a list
time_series_list = []
_d = time_series.to_dict()
for i in time_series.index:
    d = {}
    for c in time_series.columns:
        d[c] = _d[c][i]
    time_series_list.append(d)

# use a dictionary to clean data
# create the dictionary
time_series_dict = {}
for i, el in enumerate(time_series_list):
    loc = el['country_code'] + str(el['sdr_id']) + el['sdr_level']
    if loc not in time_series_dict: time_series_dict[loc] = {}
    if el['date'] not in time_series_dict[loc]: time_series_dict[loc][el['date']] = []
    time_series_dict[loc][el['date']].append(el)

## prioritizing sources and get only increasing values
def clean(reported_cases, past_cases):
    """
    reported cases: list of cases reported for a given location (of all categories)
    past_cases: dict of categories and past values
    """
    cases = {c: {} for c in past_cases}
    sources_preference = ['WHO', 'GVT', 'MINISTERE DE LA SANTE', 'UNICEF', 'ECHO'] # ordered by our preference in sources
    for el in reported_cases:
        if el['category'] != 'New cases' and el['value'] < past_cases[el['category']]: # make sure the values are increasing
            el['value'] = past_cases[el['category']]
        if not el['sources'] in cases[el['category']]: cases[el['category']][el['sources']] = []
        cases[el['category']][el['sources']].append(el)   
    for c in cases:
        if len(cases[c]):
            for s in sources_preference: # select values for the source of preference
                if s in cases[c]:
                    m = np.array([x['value'] for x in cases[c][s]]).mean()
                    cases[c] = cases[c][s][0]
                    cases[c]['value'] = m # and average over duplicate values if any
                    break
    cases = {c: cases[c] for c in cases if len(cases[c])}
    return cases
    
# build a dataframe from the clean data
clean_data = []
time_series_dict2 = {loc: {c: {} for c in time_series.category.unique()} for loc in time_series_dict}
for loc in time_series_dict:
    past_cases = {c: 0 for c in time_series.category.unique()} # initiate at 0
    # for each location, sort the data by date and clean it
    for date in sorted(time_series_dict[loc]):
        c = clean(time_series_dict[loc][date], past_cases)
        for k in c:
            clean_data.append(c[k])
            time_series_dict2[loc][k][date] = c[k]
        for k in c: past_cases[k] = c[k]['value']
                
ts_clean = pd.DataFrame(clean_data)
print 'Data size before cleaning:', len(time_series.index), 'rows. Data size after cleaning:', len(ts_clean.index), 'rows'
ts_clean.head()
```

```python
# interpolate missing data
first_day = ts_clean.date.min() - 1
interpolated_data = []

# find the day when Ebola is reported for the first time by location
ebola_start_days = {}
for loc in time_series_dict2:
    start_day = 10000000
    for c in time_series_dict2[loc]:
        if len(time_series_dict2[loc][c]):
            start_day = min(start_day, sorted(time_series_dict2[loc][c])[0])
    ebola_start_days[loc] = start_day
        
# interpolate the data
for loc in time_series_dict2:
    for c in time_series_dict2[loc]:
        last_day, last_value, yesterday_value = first_day, 0, 0
        for d in sorted(time_series_dict2[loc][c]):
            new_value = time_series_dict2[loc][c][d]['value']
            for i in range(last_day + 1, d):
                ebola_started = ebola_start_days[loc] <= i
                v = 0
                if ebola_started:  v = int(last_value + (new_value - last_value) * (i - last_day) * 1.0 / (d - last_day))
                el = time_series_dict2[loc][c][d].copy()
                el['value'] = v
                el['type'] = 'interpolate' # keep track of the origin of the data
                el['date'] = i
                el['ebola_already_reported'] = ebola_started # indicate if Ebola has already been reported for this location
                el['delta'] = v - yesterday_value
                yesterday_value = v
                interpolated_data.append(el)
                time_series_dict2[loc][c][i] = el
            ebola_started = ebola_start_days[loc] <= d
            time_series_dict2[loc][c][d]['type'] = 'original'
            time_series_dict2[loc][c][d]['ebola_already_reported'] = ebola_started
            time_series_dict2[loc][c][d]['delta'] = new_value - yesterday_value
            if d == ebola_start_days[loc]: time_series_dict2[loc][c][d]['delta'] = 0
            yesterday_value = new_value
            interpolated_data.append(time_series_dict2[loc][c][d])
            last_day, last_value = d, new_value

print 'Data size before interpolation:', len(ts_clean.index),
ts_interpolated = pd.DataFrame(interpolated_data)
print 'rows. Data size after interpolation:', len(ts_interpolated.index), 'rows'
ts_interpolated.head()
```

## A. Overview

We considered two different kinds of models to analyze our data and make predictions on the rate of growth of Ebola. One is a logistic growth model and the other one is an epidemic SIR model.  

The logistic growth model describes the disease evolution based on the assumption that the number of infected people is going to plateau in the long-run. Using this approach, we could fit a logistic growth curve to the time series dataset that contains the cumultaive case counts for each region in each country. Moreover, we could use linear regression to determine factors/covariates that explain regional variability of the growth rate variables and determine if there are important regional differences.

The epidemic SIR model describes the growth of the disease by modeling the dynamics of the disease by region using differential equations. The differential equations parameters can be found using linear regression with the reponse variable being the change in case count per day. Afterwards, regional covariates can be embeded to the model to determine which factors explain the differential equations parameters: the rate of contracting the disease and the rate of recovery/death.  

There are a couple of reasons that made us decide to choose the SIR model. The logistic growth model makes a strong assumption that the evolution of the disease is going to plateau while the SIR model does not make such an assumption. As a result, the logistic growth model is capable of giving us a general shape of how the disease evolves in the long run but it does not tell us more nuanced information as to how the disease evolves day by day. On the other hand, the SIR model has two different parameters that can tell us how fast the disease spreads and how destructive the disease is. 

# B. SIR model (description, training and testing)

$S$ stands for the non infected population, $I$ for the infected population and $R$ for the removed population (dead or immune after recovery). $N$ is the population size and $D$ the number of dead people.

For every time interval $\Delta T$:

$\Delta S = - \alpha S \frac{I}{N - D}$

$\alpha$ is the rate of contamination, $\frac{I}{N - D}$ is the density of infected people in the population and $S$ is the scale parameter

$\Delta I = \alpha S \frac{I}{N - D} - \beta I$

$\beta$ is the rate at which people die or recover and become immune.

$\Delta R = \beta I$

The number of dead $D$ is $R * \text{ mortality rate}$ and $S = N - I - R$

This model is well adapted for diseases that have an exponential growth (for small $I$, $\frac{S}{N}=\frac{N-I}{N} \sim 1$ and $\Delta I = cst  * I$)

This give us linear equations in the parameters that we are going to fit with an OLS regression. In the prediction phase, contrary to usual problems, the features $S \frac{I}{N-D}$ and $I$ are unobserved but can be constructed from the fitted values $\Delta I$ and $\Delta R$. This adds variability to the model.

```python
# function to get easily values from different dataframes
class Getter:
    def __init__(self):
        self.country_code = None
        self.sdr_id = None
        self.ts = None
        self.cov = None
        
    def get(self, country_code, sdr_id=0, attribute='Cases', date=None, delta=False):
        # keep last request in memory for faster response
        if not (country_code == self.country_code and sdr_id == self.sdr_id):
            self.country_code, self.sdr_id = country_code, sdr_id
            self.ts = ts_interpolated[(ts_interpolated.country_code == country_code) & (ts_interpolated.sdr_id == sdr_id)]
            self.cov = covariates_df[(covariates_df.country_code == country_code) & (covariates_df.sdr_id == sdr_id)]
        try:
            if attribute in self.cov:
                return self.cov.to_dict()[attribute].popitem()[1]
            else:
                df = self.ts
                if date is not None:
                    if type(date) != int and type(date) != np.int64:
                        date = (dt.datetime.strptime(date, '%Y-%m-%d') - time_base).days
                    df = self.ts[self.ts.date == date]
                if attribute in df.columns:
                    return df.to_dict()[attribute].popitem()[1] 
                else:
                    df = df[df.category == attribute]
                    if delta: k = 'delta'
                    else: k = 'value'
                    return df.to_dict()[k].popitem()[1]
        except KeyError:
            #print 'no value for', country_code, 'sdr id:', sdr_id, attribute,
            if date is not None:
                print 'date', date
                print 'query', country_code, sdr_id, attribute, date, delta
            #print
            return None
        
    def reset_cache(self):
        # reset memory
        self.country_code = None
        self.sdr_id = None
        self.ts = None
        self.cov = None
    
g = Getter()
def get(country_code, sdr_id=0, attribute='Cases', date=None, delta=False):
    return g.get(country_code, sdr_id, attribute, date, delta)
    
print 'area_sq_km:', get('GN', sdr_id=1, attribute='area_sq_km')
print 'Cases for Guinea on 2014-10-10:', get('GN', sdr_id=0, attribute='Cases', date='2014-10-10')
print 'Ebola already reported in Guinea on 2014-10-10:', get('GN', sdr_id=0, attribute='ebola_already_reported', date='2014-10-10')
```

***Create the model and target matrixes:***

```python
mortality_rate = 0.7 # from general estimates. It varies from 25 to 90%

def to_X_Y(country_code, sdr_id, train_limit):
    """convert data to OLS matrixes"""
    df = ts_interpolated[(ts_interpolated.country_code == country_code) & (ts_interpolated.sdr_id == sdr_id)]
    df = df[(df.category == 'Cases') | (df.category == 'Deaths')]
    X, Y, first_day = [], [], None
    for d in sorted(df.date.unique()):
        if d > train_limit: break
        if get(country_code, sdr_id, 'ebola_already_reported', date=d):
            if first_day is None: first_day = d
            I = get(country_code, sdr_id, 'Cases', date=d)
            if I is None:
                print d, train_limit, first_day
                raise
            D = get(country_code, sdr_id, 'Deaths', date=d)
            R = D / mortality_rate
            N = get(country_code, sdr_id, 'standardized_2014_pop')
            r = get_ratio(I, N, R)
            #D_recent = get(country_code, sdr_id, 'Recent Deaths', date=d)
            y = get(country_code, sdr_id, 'Cases', date=d, delta=True)
            x = [r, -I, 0]
            X.append(x)
            Y.append([y])
            
            y = get(country_code, sdr_id, 'Deaths', date=d, delta=True) / mortality_rate
            x = [0, I, 0]
            X.append(x)
            Y.append([y])
            
    return np.array(X), np.array(Y), first_day

def get_ratio(I, N, R):
    # return the ratio (N - I - R) / N * I
    D = R * mortality_rate
    r = I
    if N is not None: r = 1.0 * (N - I - R) * I / (N - D)
    return r
```

***Fit the model and predict:***

This is done in three steps:

- fit the parameters

- do a prediction of the variations $\Delta I$ and $\Delta D$ based on the fitted paramters and the estimated values of $I$ and $D$

- reconstruct the cumulatives variables $I$ and $D$  

The model was trained with data up to 2014-10-01, so the real prediction is from that date onward.

```python
def fit_data(country_code, sdr_id, train_limit, test_limit, days_after, should_print_table):
    """
    country_code and sdr_id identify the location
    train_limit: date until when we train
    test_limit: date until when we predict
    days_after: allow for some delay after the start of the disease before modeling (the first days usually report for past cases)
    """
    
    # I: Fit the parameters
    
    train_limit = (dt.datetime.strptime(train_limit, '%Y-%m-%d') - time_base).days
    test_limit = (dt.datetime.strptime(test_limit, '%Y-%m-%d') - time_base).days
    
    # convert the data to matrix
    X, Y, first_day = to_X_Y(country_code, sdr_id, train_limit)
    if X.shape[0] == 0:
        print 'no data before training limit (', train_limit, ')'
        return None
    
    # remove the beginning of the epidemy
    X_train = X[days_after * 2:, :]
    Y_train = Y[days_after * 2:, :]
    
    if X_train.shape[0] == 0:
        print 'no data before training limit (', train_limit, ')'
        return None
    
    # fit the OLS
    model = sm.OLS(Y_train, X_train)
    results = model.fit()
    
    # print the OLS stats
    if should_print_table: print(results.summary())
    
    # II: Do a prediction
    
    fit = X_train.dot(results.params) # for the training period, we can use the values in X_train
    
    # build the fitted times series step by step after the training period
    # get the last known data
    last_data_known = X_train[-2:, :]
    
    recent_deaths = []
    df = ts_interpolated[(ts_interpolated.country_code == country_code) & (ts_interpolated.sdr_id == sdr_id) & (ts_interpolated.category == 'Deaths')]
    for d in range(train_limit - 10, train_limit):
        _df = df[df.date == d]
        recent_deaths.append(get(country_code, sdr_id, 'Recent Deaths', date=d, delta=True))
    
    I = get(country_code, sdr_id, 'Cases', date=train_limit)
    D = get(country_code, sdr_id, 'Deaths', date=train_limit)
    N = get(country_code, sdr_id, 'standardized_2014_pop')
    R = D / mortality_rate
    
    # build the predicted data from the estimated rates and the past estimated data
    for i in range(test_limit - train_limit):
        data_interpolated = last_data_known.dot(results.params)
        fit = np.append(fit, data_interpolated)
        delta_I = data_interpolated[0]
        I += delta_I
        delta_R = data_interpolated[1]
        R += delta_R
        r = get_ratio(I, N, R)
        #recent_deaths = recent_deaths[1:]
        #recent_deaths.append(delta_D)
        last_data_known = np.array([[r, -I, 0], [0, I, 0]])
        #last_data_known[1, 2] = sum(recent_deaths)
    # get the cumulatives values from the variations
    cum_y, cum_death, delta_y, delta_death = [], [], [], []
    y, d = sum(Y[:days_after * 2:2]), sum(Y[1:days_after * 2 + 1:2]) * mortality_rate
    if type(y) != int: y, d = y[0], d[0]
    for idx, v in enumerate(fit):
        if idx % 2 == 0:
            y += v
            cum_y.append(y)
            delta_y.append(v)
        else:
            d += v * mortality_rate
            cum_death.append(d)
            delta_death.append(v)
    return cum_y, cum_death, delta_y, delta_death, range(first_day + days_after, test_limit + 1)
```

***Display the original and fitted values***

```python
def plot_fit(country_code, sdr_id, train_limit='2014-10-01', test_limit='2014-12-01', days_after=0, should_print_table=False, plot_variations=False):
    fit = fit_data(country_code, sdr_id, train_limit, test_limit, days_after, should_print_table)
    if fit is None: return
    Y_hat, D_hat, delta_Y_hat, delta_D_hat, dates = fit
    
    if not plot_variations:
        plt.plot(dates, Y_hat, label='Cases fit')
        plt.plot(dates, D_hat, label='Deaths fit')
        plot(country_code, sdr_id, ['Cases', 'Deaths'])
        plot_training_limit(train_limit)
        plt.legend(loc=2)
        plt.title('Cases and deaths: real vs model ' + country_code + '-' + str(sdr_id))
    else:
        plt.plot(dates, delta_Y_hat, label='delta Cases fit')
        plt.plot(dates, delta_D_hat, label='delta Deaths fit')
        plot(country_code, sdr_id, ['Cases', 'Deaths'], delta=True)
        plot_training_limit(train_limit)
        plt.title('Variation of cases and deaths: real vs model ' + country_code + '-' + str(sdr_id))
        plt.legend(loc=2)
    display_date()
    
def plot_training_limit(limit):
    limit = (dt.datetime.strptime(limit, '%Y-%m-%d') - time_base).days
    ax = plt.axis()
    plt.plot([limit, limit], [0, 1e50], color='r', label='training limit')
    plt.axis(ax)
    
country_code, sdr_id = 'GN', 2
plot_fit(country_code, sdr_id, days_after=30)
plot_raw(country_code, sdr_id)
```

## C. Including the covariates (description, training and testing)

We will now change our SIR model by assuming that the parameters of the differential equations depend on the regional covariates in the same way, i.e., the parameters are a function of the covariates. We will then use OLS again to fit a model to several locations simultanuously with this assumption. We decided to test this approach with Liberia, since it is the case in which we have more accurate information on the regional level.

When we incorporate the covariates, the model becomes:

$\Delta S^{(r)} = - (\sum_i \alpha_i * {cov}_i^{(r)}) S^{(r)} \frac{I^{(r)}}{N^{(r)} - D^{(r)}}$

$\Delta I^{(r)} = (\sum_i \alpha_i * {cov}_i^{(r)}) S^{(r)} \frac{I^{(r)}}{N^{(r)} - D^{(r)}} - (\sum_i \beta_i * {cov}_i^{(r)}) I^{(r)}$

$\Delta R^{(r)} = (\sum_i \beta_i * {cov}_i^{(r)}) I^{(r)}$


$r$ stands for the region indice. Note that the fitted parameters $\alpha_i$ and $\beta_i$ are independent of $r$.


$\alpha_i$ is the rate of contamination due to covariate $i$, adjusting for the other covariates.



$\beta_i$ is the rate at which people died or recovered and became immune, due to covariate $i$, after adjusting for the other covariates.

**The following code is similar to the model fitting in part III, its reading might not be extremly interesting ;)**

```python
def to_X_Y_with_covariates(country_code, sdr_ids, train_limit):
    """convert data to OLS matrixes"""
    df = ts_interpolated[(ts_interpolated.country_code == country_code) & (ts_interpolated.sdr_id.isin(sdr_ids))]
    df = df[(df.category == 'Cases') | (df.category == 'Deaths')]
    
    covariates = get_covariates(country_code, sdr_ids)
    Xs, Ys, first_days = [], [], []
    for sdr_id in sdr_ids:
        X, Y, first_day = [], [], None
        cov = covariates[sdr_id]
        for d in sorted(df.date.unique()):
            if d > train_limit: break
            if get(country_code, sdr_id, 'ebola_already_reported', date=d):
                if first_day is None: first_day = d
                I = get(country_code, sdr_id, 'Cases', date=d)
                if I is None:
                    print d, train_limit
                    raise
                D = get(country_code, sdr_id, 'Deaths', date=d)
                R = D / mortality_rate
                N = get(country_code, sdr_id, 'standardized_2014_pop')
                r = get_ratio(I, N, R)
                
                y = get(country_code, sdr_id, 'Cases', date=d, delta=True)
                x = np.append(r * cov, -I * cov)
                X.append(x)
                Y.append([y])

                y = get(country_code, sdr_id, 'Deaths', date=d, delta=True) / mortality_rate
                x = np.append(0 * cov, I * cov)
                X.append(x)
                Y.append([y])
        Xs.append(np.array(X))
        Ys.append(np.array(Y))
        first_days.append(first_day)
            
    return np.array(Xs), np.array(Ys), np.array(first_days)

def get_covariates(country_code, sdr_ids):
    # return the covariates values for the regions specified by sdr_id in the country corresponding to the country_code
    covariates = { # selected from the available covariates. They differ between countries
        'LR': ['area_sq_km', 'lb_population', 'avg_hh_size', 'sex_ratio', 'pupil_teacher_ratio', 'public_hospital', 'access_to_health_facility_5k', 'liberia_access_safe_water_2006', 'liberia_access_safe_excreta_disposal', 'num_HHS', 'restricted_access_to_market', 'standardized_2014_pop', 'pop_density', 'percent_ee', 'iwi', 'edyr', 'edyr_fem', 'edyr_male', 'urban', 'wrk_lnagr', 'wrk_unagr', 'tv', 'phone', 'electr', 'small_house', 'large_house', 'qual_floor', 'bad_floor', 'tap_water', 'bad_water', 'flush_toilet', 'bad_toilet', 'age09', 'age1019', 'age2029', 'age3039', 'age4049', 'age5059', 'age6069', 'age7079', 'age8089', 'age90hi'],
        'GN': ['area_sq_km', 'guinea_pop', 'standardized_2014_pop', 'pop_density', 'iwi', 'edyr', 'edyr_fem', 'edyr_male', 'urban', 'wrk_lnagr', 'wrk_unagr', 'tv', 'phone', 'electr', 'small_house', 'large_house', 'qual_floor', 'bad_floor', 'tap_water', 'bad_water', 'flush_toilet', 'bad_toilet', 'age09', 'age1019', 'age2029', 'age3039', 'age4049', 'age5059', 'age6069', 'age7079', 'age8089', 'age90hi'],
        'SL': ['area_sq_km', 'sl_population', 'standardized_2014_pop', 'pop_density', 'iwi', 'edyr', 'edyr_fem', 'edyr_male', 'urban', 'wrk_lnagr', 'wrk_unagr', 'tv', 'phone', 'electr', 'small_house', 'large_house', 'qual_floor', 'bad_floor', 'tap_water', 'bad_water', 'flush_toilet', 'bad_toilet', 'age09', 'age1019', 'age2029', 'age3039', 'age4049', 'age5059', 'age6069', 'age7079', 'age8089', 'age90hi']
    }
    return {sdr_id: np.array([get(country_code, sdr_id, cov_name) for cov_name in covariates[country_code]]) for sdr_id in sdr_ids}
```

```python
def fit_data_with_covariates(country_code, sdr_ids, train_limit, test_limit, days_after, should_print_table):
    """
    country_code and sdr_ids identify the location
    train_limit: date until when we train
    test_limit: date until when we predict
    days_after: allow for some delay after the start of the disease before modeling (the first days usually report for past cases)
    """
    
    # I: Fit the parameters
    
    train_limit = (dt.datetime.strptime(train_limit, '%Y-%m-%d') - time_base).days
    test_limit = (dt.datetime.strptime(test_limit, '%Y-%m-%d') - time_base).days
    
    # convert the data to matrix
    Xs, Ys, first_days = to_X_Y_with_covariates(country_code, sdr_ids, train_limit)
    Xs_train = []
    Ys_train = []
    for X, Y in zip(Xs, Ys):
        if X.shape[0] != 0:
            X_train = X[days_after * 2:, :]
            Y_train = Y[days_after * 2:, :]
            if X.shape[0] != 0:
                Xs_train.append(X_train)
                Ys_train.append(Y_train)
    
    Xs_train_stacked = np.array(np.bmat([[X_train] for X_train in Xs_train]))
    Ys_train_stacked = np.array(np.bmat([[Y_train] for Y_train in Ys_train]))
    
    # fit the OLS
    model = sm.OLS(Ys_train_stacked, Xs_train_stacked)
    results = model.fit()
    
    # print the OLS stats
    if should_print_table: print(results.summary())
    
    # II: Do a prediction
    fits = []
    covariates = get_covariates(country_code, sdr_ids)
    result = {}
    for i, sdr_id in enumerate(sdr_ids):
        X_train = Xs_train[i]
        Y = Ys[i]
        first_day = first_days[i]
        cov = covariates[sdr_id]
        fit = X_train.dot(results.params) # for the training period, we can use the values in X_train
        
    
        # build the fitted times series step by step after the training period
        # get the last known data
        last_data_known = X_train[-2:, :]

        I = get(country_code, sdr_id, 'Cases', date=train_limit)
        D = get(country_code, sdr_id, 'Deaths', date=train_limit)
        N = get(country_code, sdr_id, 'standardized_2014_pop')
        R = D / mortality_rate

        # build the predicted data from the estimated rates and the past estimated data
        for i in range(test_limit - train_limit):
            data_interpolated = last_data_known.dot(results.params)
            fit = np.append(fit, data_interpolated)
            delta_I = data_interpolated[0]
            I += delta_I
            delta_R = data_interpolated[1]
            R += delta_R
            r = get_ratio(I, N, R)
            last_data_known = np.array([np.append(r * cov, -I * cov), np.append(0 * cov, I * cov)])
        # get the cumulatives values from the variations
        cum_y, cum_death, delta_y, delta_death = [], [], [], []
        y, d = sum(Y[:days_after * 2:2]), sum(Y[1:days_after * 2 + 1:2]) * mortality_rate
        if type(y) != int: y, d = y[0], d[0]
        for idx, v in enumerate(fit):
            if idx % 2 == 0:
                y += v
                cum_y.append(y)
                delta_y.append(v)
            else:
                d += v * mortality_rate
                cum_death.append(d)
                delta_death.append(v)
        result[sdr_id] = [cum_y, cum_death, delta_y, delta_death, range(first_day + days_after, test_limit + 1)]
    return result
```

```python
def plot_fit_with_covariates(country_code, sdr_ids, train_limit='2014-10-15', test_limit='2014-12-01', days_after=0, should_print_table=False):
    fits = fit_data_with_covariates(country_code, sdr_ids, train_limit, test_limit, days_after, should_print_table)
    for sdr_id in sdr_ids:
        fit = fits[sdr_id]
        if fit is None: return
        
        name_region = data[0][data[0].id == sdr_id]['gn_name']
        name = name_region[name_region.index[0]]
        
        Y_hat, D_hat, delta_Y_hat, delta_D_hat, dates = fit

        plt.plot(dates, Y_hat, label='Cases fit')
        plt.plot(dates, D_hat, label='Deaths fit')
        plot(country_code, sdr_id, ['Cases', 'Deaths'])
        plot_training_limit(train_limit)
        plt.legend(loc=2)
        plt.title('Cases and deaths ' + name)
        display_date()
        plt.show()
```

This makes us realize that the SIR model implicitly assumes an exponential growth of the number of cases/number of deaths, which works reasonably well on national level but not that well on more local levels. # D. Conclusion

After training and predicting with several models, we conclude that an SIR model is adequate to model disease growth on the national level, since it seems that at that level of aggregation, the implicit assumption of exponential growth (that the SIR makes) seems to hold. However, when deaggregating the data regionaly, disease growth shape varies greatly (e.g. sigmoidal curves, or curves that grow slower than an exponential function), which means that more flexibility needs to be built into the model.

We still consider once this flexibility is incorporated, valuable insights could be obtained by fitting a global model to all regions (assuming the parameters of the growth of the disease are explained as the same function of the covariates).
